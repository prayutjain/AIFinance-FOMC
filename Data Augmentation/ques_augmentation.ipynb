```python
import openai

openai.api_key = ''

def augment_data(texts, num_augmentations):
    augmented_data = []

    for text in texts:
        augmentations = [text]

        # Augmentation 1: Text Expansion
        expansion_prompt = "Expand the following text: \"" + text + "\""
        expansion_response = openai.Completion.create(
            engine='text-davinci-003',
            prompt=expansion_prompt,
            max_tokens=300,
            temperature=0.8,
            n=num_augmentations[0],
            stop=None
        )
        expansion_texts = [choice['text'].strip() for choice in expansion_response.choices]
        augmentations.extend(expansion_texts[:num_augmentations[0]])  # Append the expanded texts

        # Augmentation 2: Paraphrasing
        paraphrase_prompt = "Paraphrase the following text: \"" + text + "\""
        paraphrase_response = openai.Completion.create(
            engine='text-davinci-003',
            prompt=paraphrase_prompt,
            max_tokens=80,
            temperature=0.8,
            n=num_augmentations[1],
            stop=None
        )
        paraphrase_texts = [choice['text'].strip() for choice in paraphrase_response.choices]
        augmentations.extend(paraphrase_texts[:num_augmentations[1]])  # Append the paraphrased texts

        # Augmentation 3: Style Transfer
        style_transfer_prompt = "Transfer the style of the following text: \"" + text + "\" to an informal writing style. Ensure that the content remains intact."
        style_transfer_response = openai.Completion.create(
            engine='text-davinci-003',
            prompt=style_transfer_prompt,
            max_tokens=150,
            temperature=0.8,
            n=num_augmentations[2],
            stop=None
        )
        style_transfer_texts = [choice['text'].strip() for choice in style_transfer_response.choices]
        augmentations.extend(style_transfer_texts[:num_augmentations[2]])  # Append the style transferred texts

        # Augmentation 4: Problem-Solution Prompt
        prob_prompt = "Identify a problem in the following text: \"" + text + "\"  Propose effective solutions to address the problem in the form of a question. Do not change the meaning of the text."
        prob_response = openai.Completion.create(
            engine='text-davinci-003',
            prompt=prob_prompt,
            max_tokens=200,
            temperature=0.8,
            n=num_augmentations[3],
            stop=None
        )
        prob_texts = [choice['text'].strip() for choice in prob_response.choices]
        augmentations.extend(prob_texts[:num_augmentations[3]])

        augmented_data.append(augmentations)

    return augmented_data

original_texts = [ques1,ques2,ques3,ques4,ques5,ques6,ques7,ques8,ques9,ques10,ques11,ques12,ques13,ques14,ques15,ques16,ques17,ques18,ques19,ques20,ques21,ques22,ques23,ques24]
num_augmentations = [2, 3, 2, 3]  # Specify the desired number of augmentations for each type

augmented_data = augment_data(original_texts, num_augmentations)

for augmented_texts in augmented_data:
    print("Original text:")
    print(augmented_texts[0])
    print("Augmentations:")
    for i, augmentation in enumerate(augmented_texts[1:]):
        print(f"{i+1}. {augmentation}")

```

```python
df = pd.DataFrame(augmented_data, columns=["Original Text"] + [f"Augmentation {i+1}" for i in range(sum(num_augmentations))])
```
